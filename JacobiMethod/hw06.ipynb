```python
# 繪製圖形
import matplotlib.pyplot as plt
from sympy import *
from sympy.plotting.plot import MatplotlibBackend, Plot
from ast import literal_eval
import random

file=open('data.out', mode='r')
word=file.readline()
str1=word.split(' ')
M=literal_eval(str1[0])
offDiag=[]
x=[]
for i in range(M):
    word=file.readline()
    str1=word.split(' ')
    offDiag.append(literal_eval(str1[0]))
    x.append(i+1)
eigenvalues=file.readline()
file.close()
print(eigenvalues)
plt.figure(figsize=(10,8),linewidth = 1)
plt.plot(x,offDiag)
plt.title("Jacobian method",fontsize=15)
plt.legend(loc = "best", fontsize=10)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.xlabel("iteration(k)", fontsize=15)
plt.ylabel("offDiag[k]", fontsize=15)
plt.savefig('iter_offDiag_N_20.png')
```

```python
# 繪製圖形
import matplotlib.pyplot as plt
from sympy import *
from sympy.plotting.plot import MatplotlibBackend, Plot
from ast import literal_eval
import random

def get_sympy_subplots(plot: Plot):
    backend = MatplotlibBackend(plot)

    backend.process_series()
    backend.fig.tight_layout()
    return backend.fig, backend.ax[0]

file=open('data_NM.out', mode='r')
words=file.readlines()
file.close()
M=[]
N=[]
Y=[]
cur=4
for word in words:
    str1=word.split(' ')
    M.append(literal_eval(str1[0]))
    N.append(cur)
    Y.append(cur*cur)
    cur+=1
plt.figure(figsize=(10,8),linewidth = 1)
plt.plot(N,M,'.')
plt.plot(N,Y,label='Y=X^2')
plt.title("Jacobian method",fontsize=15)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.xlabel("N", fontsize=15)
plt.ylabel("M", fontsize=15)
plt.legend(loc = "best", fontsize=10)
plt.savefig('NM_iter.png')
```

```python
# 繪製圖形
import matplotlib.pyplot as plt
from sympy import *
from sympy.plotting.plot import MatplotlibBackend, Plot
from ast import literal_eval
import random

file=open('data_delta10.out', mode='r')
words=file.readlines()
file.close()
delta=[]
ratio=[]
cur=0
for word in words:
    str1=word.split(' ')
    ratio.append(literal_eval(str1[0]))
    #str2="δ"+str(cur+1)+"/δ0"
    delta.append(cur)
    cur+=1

plt.figure(figsize=(10,8),linewidth = 1)
plt.plot(delta,ratio,'.')
plt.title("Jacobian method(n = 10)",fontsize=15)
plt.xticks(fontsize=8)
plt.yticks(fontsize=10)
plt.xlabel("i", fontsize=15)
plt.ylabel("δi/δ0", fontsize=15)
plt.savefig('delta_10.png')
```
