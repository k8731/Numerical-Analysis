```python
# 讀取資料
from ast import literal_eval
real=[]
real2=[]
cell_1=[[],[]]
cell_4=[[],[]]
cell_16=[[],[]]
cell_64=[[],[]]
N_2=[[],[]]
N_3=[[],[]]
N_4=[[],[]]
value=[]
error=[]
file=open('data.out', mode='r')
words=file.readline()
str1=words.split(' ')
real.append(literal_eval(str1[2]))
real.append(literal_eval(str1[2]))
real.append(literal_eval(str1[2]))
real2.append(literal_eval(str1[2]))
real2.append(literal_eval(str1[2]))
real2.append(literal_eval(str1[2]))
real2.append(literal_eval(str1[2]))
words=file.readlines()
for word in words:
    str1=word.split(' ')
    value.append(literal_eval(str1[6]))
    error.append(literal_eval(str1[10]))
for i in range(12):
    error.append(abs(value[i]-real[0]))
    if i%3==0:
        N_2[0].append(value[i])
        N_2[1].append(error[i])
    elif i%3==1:
        N_3[0].append(value[i])
        N_3[1].append(error[i])
    else:
        N_4[0].append(value[i])
        N_4[1].append(error[i])
    
    if(i<3):
        cell_1[0].append(value[i])
        cell_1[1].append(error[i])
    elif(i<6):
        cell_4[0].append(value[i])
        cell_4[1].append(error[i])
    elif(i<9):
        cell_16[0].append(value[i])
        cell_16[1].append(error[i])
    else:
        cell_64[0].append(value[i])
        cell_64[1].append(error[i])
```

```python
import matplotlib.pyplot as plt
N = ['2x2','3x3','4x4']
plt.figure(figsize=(15,10),linewidth = 2)
plt.plot(N,real,'s-',color = 'black', label="real value")
plt.plot(N,cell_1[0],'s-',color = 'r', label="cell=1*1")
plt.plot(N,cell_4[0],'s-',color = 'g', label="cell=2*2")
plt.plot(N,cell_16[0],'s-',color = 'b', label="cell=4*4")
plt.plot(N,cell_64[0],'s-',color = 'purple', label="cell=8*8")
plt.title("Gaussian Quadrature vs Actual Integral",fontsize=20)
plt.legend(loc = "best", fontsize=20)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.xlabel("N", fontsize=30)
plt.ylabel("value", fontsize=30)
plt.savefig('comparison_chart.png')
```
