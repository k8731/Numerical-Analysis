```python
# 繪製圖形
import matplotlib.pyplot as plt
from sympy import *
from sympy.plotting.plot import MatplotlibBackend, Plot
from ast import literal_eval
import random
n=[]
for i in range(10,110,10):
    n.append(i)
w=[1.0,1.1,1.2,1.3,1.4,1.5,1.6,1.7,1.8,1.9]
e=[]

file=open('data1.out', mode='r')
for i in range(2,41,2):
    word=file.readline()
    tp=[]
    for j in range(10):
        word=file.readline()
        str1=word.split(' ')
        tp.append(literal_eval(str1[1]))
    e.append(tp)
file.close()
plt.figure(figsize=(10,8),linewidth = 1)
cnt=2
for ee in e:
    c=(random.random(),random.random(),random.random()) # 每個initial point有一個隨機color
    plt.plot(w,ee,'s-',color = c,label='n = '+str(cnt))
    cnt+=2
plt.title("SOR",fontsize=15)
plt.legend(loc = "best", fontsize=10)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.xlabel("ω", fontsize=15)
plt.ylabel("iteration", fontsize=15)
plt.savefig('SOR1.png')
```
