```python
# 
import matplotlib.pyplot as plt
from sympy import *
from sympy.plotting.plot import MatplotlibBackend, Plot
from ast import literal_eval
import random

file=open('data3.out', mode='r')
words=file.readline()
words=file.readlines()
x=[]
y=[]
yt=[]
ori=20
end=51
cnt=ori
for i in range(ori,end):
    x.append(i)
for w in words:
    str1=w.split(' ')
    yt.append(literal_eval(str1[7]))
    cnt+=1
    if(cnt==end):
        y.append(yt)
        yt=[]
        cnt=ori
    
file.close()
plt.figure(figsize=(10,8),linewidth = 1)
cnt=8
for yt in y:
    c=(random.random(),random.random(),random.random()) # initial pointcolor
    plt.plot(x,yt,'s-',color = c,label='n = '+str(cnt))
    cnt+=1
plt.title("Gaussian elimation",fontsize=15)
plt.legend(loc = "best", fontsize=10)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.xlabel("N", fontsize=15)
plt.ylabel("value(log)", fontsize=15)
plt.savefig('Gaussian_elimation3.png')
```

```python
# 
import matplotlib.pyplot as plt
from sympy import *
from sympy.plotting.plot import MatplotlibBackend, Plot
from ast import literal_eval
import random

file=open('data8.out', mode='r')
x=[]
y=[[],[],[]]
for j in range(4,21):
    for i in range(3):
        w=file.readline()
        str1=w.split(' ')
        y[i].append(literal_eval(str1[8]))
for i in range(4,21):
    x.append(i)

file.close()
plt.figure(figsize=(10,8),linewidth = 1)
cnt=0
for yt in y:
    cnt+=1
    c=(random.random(),random.random(),random.random()) # initial pointcolor
    plt.plot(x,yt,'s-',color = c,label='method = '+str(cnt))
plt.title("Compare with 3 method",fontsize=15)
plt.legend(loc = "best", fontsize=10)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.xlabel("n", fontsize=15)
plt.ylabel("value(log10)", fontsize=15)
plt.savefig('Compare1.png')
```
